```python
batch_size = 5
sample_df = pd.read_csv('/work/ptyagi/masterthesis/tmp/annotated_tweets_with_replies.csv')
```

```python
#sample_df['is_annotated'] = False 
#sample_df['manual_emotion_label'] = ''
#sample_df.to_csv('/work/ptyagi/masterthesis/tmp/annotated_tweets_with_replies.csv', index=False)
```

```python
start_idx = sample_df[sample_df['is_annotated'] == False].index[0]
start_idx
```

```python
from IPython.display import clear_output

for i in range(start_idx, len(sample_df), batch_size):
    for j in range(i, min(i + batch_size, len(sample_df))):
        if not sample_df.iloc[j]['is_annotated']:

            clear_output(wait=True)

            print(f"Tweet: {sample_df.iloc[j]['tweet_text']}")
            print(f"Reply: {sample_df.iloc[j]['replies']}")

            manual_label = input("Enter the emotion label: ")

            sample_df.at[sample_df.index[j], 'manual_emotion_label'] = manual_label
            sample_df.at[sample_df.index[j], 'is_annotated'] = True

    sample_df.to_csv('/work/ptyagi/masterthesis/tmp/annotated_tweets_with_replies.csv', index=False)
    print(f"Batch of {batch_size} samples annotated and saved.")
```
